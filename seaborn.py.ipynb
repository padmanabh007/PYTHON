```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
#Corelation with heatmap
df.corr()#where the values must be always between -1 to 1 for finding the coorelation the values should fe floating or integer type
```

```python
df.dtypes# to find the datatype
```

```python
sns.heatmap(df.corr())# to visulatitze the corellation
```

```python
#JOIN PLOT it helps to do the univarient analysis but should have two features
sns.jointplot('tip','total_bill',df,'reg')#were kind represent what kind of data you required as hexegon we use hex and for regression use reg
```

```python
#pair plotis used if we have more than 2 features is is also called scatter plot
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')#where its is drawn in the bassis of sex
```

```python
#DIST PLOT is used to create a histogram
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)#noew y axis will show the count
```

```python
#CATEGORICAL PLOT
#COUNT PLOT
sns.countplot('sex',data=df)#it will plot on the basis of sex based on the count by eitherin xor in y axis 
```

```python
#BAR PLOT it is drawn the xas well as y axis is based on cateconical feature
sns.barplot(x='total_bill',y='sex',data=df) 
```

```python
#BOX PLOT is based on the percentile
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')#palette will give rainbow colour otherwise default colour
```

```python
sns.boxplot(data=df,orient='v')#give the box plot on all the data values and in vertical orientation
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)#in which the diagram is ploted on the basis of th eplotter
```

```python
#VIOLIN PLOT helps to see both in kernal as well as box plot
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')#the internally is the boxplot 
```
